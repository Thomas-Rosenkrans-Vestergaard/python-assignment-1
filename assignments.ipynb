# Write a program, which downloads a set of files from the internet.

```python
import urllib.request
import re

def download(urls):
    for url in urls:
        print("fetching " + url)
        contents = urllib.request.urlopen(url).read()
        name = re.search("([a-zA-Z0-9-_\.]+)$", url).group().replace('.txt', '') + ".txt"
        print("writing ./" + name)
        f = open(name, 'w')
        f.write(str(contents))
        f.close()
```

```python
download(["https://facebook.com", "https://youtube.com", "https://tvestergaard.com", "https://stackoverflow.com/questions/37372603/how-to-remove-specific-substrings-from-a-set-of-strings-in-python"])
```

# Write a program that converts the Excel spreadsheet ./iris_data.xlsx into a CSV file with the same data.

```python
from openpyxl import load_workbook
import csv

def xlsx_to_csv(read, write):
    workbook = load_workbook(read)
    sheet    = workbook[workbook.sheetnames[0]]
    with open(write, 'w', newline="") as output:
        writer = csv.writer(output, delimiter=',', quotechar='"')
        for row in sheet.iter_rows():
            writer.writerow(list(map(lambda c: c.value, row)))

def read_csv(f):
    with open(f) as input_file:
        reader = csv.reader(input_file, delimiter=',')
        for row in reader:
            print(row)
            
xlsx_to_csv('iris_data.xlsx', 'iris_data.csv')
read_csv('iris_data.csv')
```

# Write a function that reads the befkbhalderstatkode.csv

```python
import json
from collections import defaultdict

long_file = "befkbhalderstatkode.csv"        # the original file
short_file = "befkbhalderstatkode_short.csv" # a file that wont crash jupyter notebook (100 records)

# Creates the structure using defaultdicts
def create_structure_dd(file):

    def nested_dd():
        return defaultdict(nested_dd)

    structure = nested_dd()
    with open(file) as f:
        lines = f.readlines()
        for x in range(1, len(lines)):
            values = list(map(int, lines[x].split(",")))
            structure[values[0]][values[1]][values[2]][values[3]] = values[4]
            
    return structure

# Prints a structure using json
def print_structure(structure):
    print(json.dumps(structure, indent=4))

structure = create_structure_dd(short_file)
print_structure(structure)
```
